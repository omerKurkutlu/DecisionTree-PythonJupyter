```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("data.csv")    #read csv file
```

```python
inputs=df.drop(columns=['Buys'])      # here get input columns from file
outputs=df['Buys']                    #here get output columns from file
```

```python
le = LabelEncoder()          # create instance of label encoder
```

```python
inputs['Income_Encoded']=le.fit_transform(inputs['Income'])                #encode Income Column
inputs['Gender_Encoded']=le.fit_transform(inputs['Gender'])                #encode Gender Column
inputs['MarialStatus_Encoded']=le.fit_transform(inputs['Marial Status'])   #encode Marial Status Column
```

```python
inputs = inputs.drop(columns=['ID','Income','Gender','Marial Status'])     #drop the un-encoded columns and ID
inputs
```

```python
model=DecisionTreeClassifier()          # create decision tree model
model.fit(inputs,outputs)               #assign input output data to model
```

```python
model.score(inputs,outputs)
```

```python
prediction=model.predict([[22,0,0,1]])   # we areprdiction for different input
prediction
```
